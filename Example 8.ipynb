```python
from geodesy import dms2rad, rad2dms, deg2rad, geod1

# GRS80 ellipsoid
a = 6378137    # [meter]
f = 1/298.257222101
b = a*(1 - f)

# Launch site
latAn = dms2rad([69, 17, 46])
lonAn = dms2rad([16, 6, 29])

az1 = deg2rad(87.9)
d = 550e3

[latX, lonX, az2] = geod1(a, b, latAn, lonAn, az1, d)

d, m, s = rad2dms(latX)
print("latX: {:3d} {:2d} {:2.0f} N".format(d, m, s))

d, m, s = rad2dms(lonX)
print("lonX: {:3d} {:2d} {:2.0f} E".format(d, m, s))
```
