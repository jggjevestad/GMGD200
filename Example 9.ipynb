```python
from geodesy import pi, dms2rad, rad2deg, geod2

# GRS80 ellipsoid
a = 6378137    # [meter]
f = 1/298.257222101
b = a*(1 - f)

# Copenhagen
latCo = dms2rad([55, 41, 0])
lonCo = dms2rad([12, 34, 0])

# Anchorage
latAn = dms2rad([61, 12, 0])
lonAn = -dms2rad([149, 53, 0])  # West!

[az_Co_An, az_An_Co, d_Co_An] = geod2(a, b, latCo, lonCo, latAn, lonAn)

print("d_Co_An: {:8.1f} km".format(d_Co_An/1e3))
```

```python
print("az_Co_An: {:5.1f} deg".format(rad2deg(az_Co_An)))
```

```python
print("az_An_Co: {:5.1f} deg".format(rad2deg(az_An_Co + pi)))
```
