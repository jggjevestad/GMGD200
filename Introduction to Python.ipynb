```python
# Library "geodesi.py"
from numpy import pi

# Degree to radian
def deg2rad(deg):
    return deg*(pi/180)

# Radian to degree
def rad2deg(rad):
    return rad*(180/pi)

# Degree to gradian
def deg2grad(deg):
    return deg*(200/180)

# Gradian to degree
def grad2deg(grad):
    return grad*(180/200)

# Gradian to radian
def grad2rad(grad):
    deg = grad2deg(grad)
    rad = deg2rad(deg)
    return rad
```

```python
# Test
grad = 100
rad = grad2rad(grad)
print(rad)
```
